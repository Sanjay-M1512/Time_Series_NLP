```python
import pandas as pd
import os
import sys
import warnings

warnings.filterwarnings("ignore")

def load_csv(filename):
    path = os.path.join(sys.path[0], filename)
    if not os.path.exists(path):
        print(f"Error: File '{filename}' not found in script directory!")
        sys.exit()
    return pd.read_csv(path)

def main():
    filename = input().strip()
    df = load_csv(filename)

    # First 5 rows
    print("First 5 rows of the dataset:")
    print(df.head())

    # Missing values (EXCLUDE Date column)
    print("\nMissing values in dataset:")
    print(df.drop(columns=["Date"]).isnull().sum())

    # Duplicate rows
    print("\nNumber of duplicate rows:",df.duplicated().sum())

    # Close price statistics
    print("\nClose price summary statistics:")
    print(df["Close"].describe())

if __name__ == "__main__":
    main()
```
