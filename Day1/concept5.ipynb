```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv(
    r"C:\Users\HP\OneDrive\Desktop\NLP\Data\ML471_S1_Datafile_Concept.csv"
)

# Convert DATE column to datetime
df['DATE'] = pd.to_datetime(df['DATE'])

# Set DATE as index
df.set_index('DATE', inplace=True)

# First-order differencing
df['Differenced_Consumption'] = df['Consumption'].diff()
```

```python
plt.figure(figsize=(12, 5))
plt.plot(df.index, df['Differenced_Consumption'])
plt.title("Differenced Power Consumption")
plt.xlabel("Year")
plt.ylabel("Differenced Power Consumption")
plt.grid(True)
plt.show()
```
