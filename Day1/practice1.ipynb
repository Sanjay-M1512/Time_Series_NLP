```python
import pandas as pd
import os
import sys

def load_csv(filename):
    path = os.path.join(sys.path[0], filename)
    if not os.path.exists(path):
        print(f"Error: File '{filename}' not found in script directory!")
        sys.exit()
    return pd.read_csv(path)

file_name = input()

df = load_csv(file_name)

df['Date'] = pd.to_datetime(df['Date'])
df.set_index('Date', inplace=True)

print("First 5 rows of the dataset:")
first_five = df.reset_index().head()
first_five['Date'] = first_five['Date'].dt.strftime('%-m/%-d/%Y')
print(first_five)

df = df.ffill()

print("\nMissing values in dataset:")
print(df.isnull().sum())

duplicate_count = df.duplicated().sum()
print("\nNumber of duplicate rows:", duplicate_count)

df = df.drop_duplicates()

print("\nClose price summary statistics:")
print(df['Close'].describe())
```
