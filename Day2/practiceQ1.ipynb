```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from statsmodels.graphics.tsaplots import plot_acf, plot_pacf
```

```python
df = pd.read_csv(r"C:\Users\HP\OneDrive\Desktop\NLP\Data\ML471_S2_Datafile_Practice.csv")

df['Date'] = pd.to_datetime(df['Date'])

df = df.sort_values('Date')

df.set_index('Date', inplace=True)

series = df['Close_diff'].dropna()

df = df.drop(columns=['SMA_10', 'SMA_30', 'SES'], errors='ignore')
```

```python
plt.figure(figsize=(12,6))
plt.subplot(1,2,1)
plot_acf(series, ax=plt.gca(), color='blue')
plt.title('ACF of Differenced Date')

plt.subplot(1,2,2)
plot_pacf(series, ax=plt.gca(), method = 'ywm' ,color='red')
plt.title('PACF of Differenced Date')

plt.tight_layout()
plt.show()
```
