```python
import pandas as pd
import numpy as np
import os, sys
from sklearn.model_selection import train_test_split as tts

df = pd.read_csv(os.path.join(sys.path[0], input()))

print("Dataset Preview:")
print(df.head())
print()

print("Dataset Information:")
print(df.info())
print()

df.set_index("Datetime", inplace = True)
print("Missing Value Check:")
print(df.isna().sum())
df.dropna(inplace = True)
print("After missing value handling:")
print(df.isna().sum())
print()

print("ACF and PACF Analysis:")
print("Time series module not available. Skipping ACF/PACF plots.")
print()

tr, te = tts(df['Consumption'], test_size = 0.2, shuffle = False)

print("Train-Test Split:")
print(f"Training records: {len(tr)}")
print(f"Testing records: {len(te)}")
```
