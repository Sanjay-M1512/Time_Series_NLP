```python
import pandas as pd
import matplotlib.pyplot as plt
from statsmodels.tsa.holtwinters import ExponentialSmoothing
```

```python
file_path = r"C:\Users\HP\OneDrive\Desktop\NLP\Data\ML471_S4_Datafile_Concept.csv"
df = pd.read_csv(file_path)

df['Datetime'] = pd.to_datetime(df['Datetime'])
df.set_index('Datetime', inplace=True)

ts = df['Consumption']
```

```python
model = ExponentialSmoothing(
    ts,
    trend='add',
    seasonal='add',
    seasonal_periods=12
)

fit_model = model.fit()

smoothed = fit_model.fittedvalues
```

```python
plt.figure(figsize=(14,6))
plt.plot(ts, label='Original Data', color='blue')
plt.plot(smoothed, label='Final Smoothed Data', color='red')
plt.title("Final Exponential Smoothing")
plt.xlabel("Year")
plt.ylabel("Electricity Consumption")
plt.legend()
plt.grid(True)
plt.show()
```
