```python
import os 
os.environ["TF_CPP_MIN_LOG_LEVEL"]='3'

import sys,spacy

nlp=spacy.load('en_core_web_sm')
fn=input()
print("Enter text file name:")
with open(os.path.join(sys.path[0],fn),'r',encoding='utf-8') as f:
    c=f.read()

print("=== Original Text Sample (First 300 chars) ===")

print(c[:300])
ccc=nlp(c)

print("=== Text After Stop Word Removal (Sample) ===")

t=[cc.text for cc in ccc if not cc.is_stop and not cc.is_space]

print(" ".join(t[:50]))

cs={'set', 'example', 'whenever', 'whatever'}


print("Custom stop words added:",cs)

for w in cs:
    nlp.Defaults.stop_words.add(w)

print("Is 'example' a stop word?","example" in nlp.Defaults.stop_words)
print("Total Stop Words Now:",len(nlp.Defaults.stop_words))

nlp.Defaults.stop_words.remove("example")
print("Removed stop word: {'example'}")
print("Is 'example' a stop word now?","example" in nlp.Defaults.stop_words)
print("Total Stop Words After Removal:",len(nlp.Defaults.stop_words))

print("=== Stop Word Removal Examples ===")
a="The quick brown fox jumps over the lazy dog."
print("Input 1:",a)
c1=nlp(a)
t1=[cc.text for cc in c1 if not cc.is_stop and not cc.is_space]
print("After Stop Word Removal:"," ".join(t1))

print("Input 2: I am learning Natural Language Processing with Python.")
a="I am learning Natural Language Processing with Python."
c1=nlp(a)
t1=[cc.text for cc in c1 if not cc.is_stop and not cc.is_space]
print("After Stop Word Removal:"," ".join(t1))

print("Input 3: What is the best way to learn natural language processing?")
a="What is the best way to learn natural language processing?"
c1=nlp(a)
t1=[cc.text for cc in c1 if not cc.is_stop and not cc.is_space]
print("After Stop Word Removal:"," ".join(t1))


```
