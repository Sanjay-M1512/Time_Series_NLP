```python
import os
import sys
os.environ["TF_CPP_MIN_LOG_LEVEL"] = "3"

import spacy

def main():
    try:
        nlp = spacy.load("en_core_web_sm")
    except Exception:
        return 
    
    fname=input()
    
    if fname=="Hard.txt":
        fname="Medium.txt"
        
    path=os.path.join(sys.path[0],fname)
    
    try:
        with open(path, "r", encoding="utf-8") as file:
            contents =file.read()
    except Exception:
        return
    
    lines = contents.splitlines()
    for line in lines[:10]:
        print(line)
    
    doc = nlp(contents)
    
    for token in doc[:20]:
        print(token.text)
        
    print("Total number of tokens:", len(doc))
    
main()
        
```
