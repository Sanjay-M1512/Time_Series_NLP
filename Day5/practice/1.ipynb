```python
import os
os.environ['TF_CPP_MIN_LOG_LEVEL'] = '3'
import sys, spacy, warnings as w
w.filterwarnings("ignore")

nlp = spacy.load("en_core_web_sm")

with open(os.path.join(sys.path[0], input()), encoding = 'utf-8') as f:
    text = f.read()
    
doc = nlp(text)
print("First 10 lines from the file:")
for i in text.splitlines()[:10]:
    print(i)
print()

print("First 20 tokens:")
print([i.text for i in doc][:20])
```
