```python
import os
os.environ['TF_CPP_MIN_LOG_LEVEL']='2'
import sys,spacy
nlp=spacy.load('en_core_web_sm')
import warnings as w
w.filterwarnings("ignore")

with open(os.path.join(sys.path[0],input()),encoding='utf-8') as f:
    c=f.read()

print("First 10 lines from the file:")
for i in c.splitlines()[:10]:
    print(i)
    
print("First 20 tokens:")
d=nlp(c)
t=[a.text for a in d[:20]]
print(t)

print("POS Tagging Output:")
print("Word     POS  Tag")
print("-"*30)

for i in d:
    print(i.text,i.pos_,i.tag_)
```
