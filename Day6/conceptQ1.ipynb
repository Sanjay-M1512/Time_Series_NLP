```python
import spacy
from spacy import displacy

npl = spacy.load("en_core_web_sm")
filepath = r"C:\Users\HP\OneDrive\Desktop\NLP\Data\ML471_S7_Datafile_Concept.txt"

with open(filepath, 'r', encoding='utf-8') as file:
    text = file.read( )

doc = npl(text)

print("TOKEN\t\tDEP\t\tHEAD")
print("-" * 50)
for token in doc:
    print(f"{token.text:<12}\t\t{token.dep_:<10}\t\t{token.head.text}")

sentences = list(doc.sents)
for i, sent in enumerate(sentences[:5], 1):
    print(f"\nSentence {i}: {sent.text}\n")    
    displacy.render(sent, style="dep", jupyter=True, options={"distance":90})
    
```
